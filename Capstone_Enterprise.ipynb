## **Penanganan Outliers**

### **Pengecekkan Outliers**

```python
kolom = ['Debit', 'Kredit', 'Saldo']

results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

Persentase outliers yang tinggi terletak pada kolom "Kredit" besar 13%, "Saldo" sebesar 12% menunjukkan bahwa ada sejumlah besar data yang memiliki nilai yang jauh dari nilai-nilai lain dalam kolom tersebut. Sedangkan persentase outliers yang lebih rendah terletak pada kolom "Debit" sebesar 5% menunjukkan bahwa proporsi data yang dianggap sebagai pencilan dalam kolom tersebut relatif lebih kecil dibandingkan dengan kolom lainnya.

```python
Q1 = df['Saldo'].quantile(0.25)
Q3 = df['Saldo'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Saldo'] >= lower_bound) & (df['Saldo'] <= upper_bound)]
df.head()
```

Menggunakan metode IQR (Interquartile Range) untuk menentukan batas bawah dan batas atas, dan kemudian memfilter data untuk menyaring nilai-nilai di luar rentang tersebut.

Melakukan proses normalisasi nilai-nilai pada data yang telah dipilih sebelumnya (x), Hasilnya, x_scaled akan berisi data yang telah di-transformasikan dengan penskalaan yang telah ditentukan oleh StandardScaler sesuai dengan data aslinya. Ini berfungsi untuk mencegah fitur-fitur dengan rentang nilai yang besar memberikan pengaruh yang lebih besar dalam algoritma mechine learning.

```python
inertia_values = []
k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Jumlah Kluster (K)')
plt.ylabel('Inertia')
plt.title('Metode Elbow untuk K Optimal')
```

Visualisasi metode Elbow, ini menentukan jumlah klaster yang optimal untuk data. Hal ini dilakukan dengan memplot jumlah inersia atau distorsi dalam klaster terhadap jumlah klaster yang berbeda. Inersia biasanya diplot pada sumbu y sedangkan jumlah klaster diplot pada sumbu x. Hasil dari Visualisasi Elbow diatas ialah cluster yang optimal pada model ini adalah cluster 3.

```python
pipeline_kmeans = Pipeline([
    ('scaler', StandardScaler()),
    ('kmeans', KMeans(n_clusters=3, random_state=42))
])
```

## **Review Process**

### **Cross Validation**

```python
iterasi = 5
skor_silhouette_kmeans = [0.534411]
skor_silhouette_hc = [0.453829]

data = np.random.rand(100, 3)

for i in range(iterasi):
  data_sample = resample(data, random_state=i)
  scaler = StandardScaler()
  data_sample_scaled = scaler.fit_transform(data_sample)

  kmeans = KMeans(n_clusters=4, random_state=42)
  kmeans.fit(data_sample_scaled)
  skor_silhouette_kmeans.append(silhouette_score(data_sample_scaled, kmeans.labels_))

  hc = AgglomerativeClustering(n_clusters=4)
  cluster_labels_hc = hc.fit_predict(data_sample_scaled)
  skor_silhouette_hc.append(silhouette_score(data_sample_scaled, cluster_labels_hc))

plt.figure(figsize=(10, 6))
plt.plot(skor_silhouette_kmeans, marker='o', label='KMeans')
plt.plot(skor_silhouette_hc, marker='x', label='Hierarchical Clustering')
plt.title('Cross Validation')
plt.xlabel('Iterasi')
plt.ylabel('Skor Silhouette')
plt.legend()
plt.grid(True)
plt.xticks(range(iterasi))
plt.show()
```

Menunjukkan hasil dari cross-validation untuk dua algoritma klasterisasi, yaitu KMeans dengan warna garis biru dan Hierarchical Clustering dengan warna garis orange, dengan metrik evaluasi Silhouette Score. Grafik menunjukkan bagaimana nilai Silhouette Score berubah seiring dengan jumlah iterasi dalam proses cross-validation.

Hasilnya ialah

- Hasil dari grafik menunjukkan KMeans secara konsisten memiliki kinerja yang lebih baik daripada Hierarchical Clustering selama proses cross-validation.

- KMeans cenderung memberikan nilai Silhouette Score yang lebih tinggi secara keseluruhan. Klaster yang dihasilkan oleh KMeans lebih berkualitas dalam hal pemisahan antar klaster.

```python
# Tampilkan hasil evaluasi
avg_cv_scores = {'KMeans': skor_silhouette_kmeans[1:], 'Hierarchical Clustering': skor_silhouette_hc[1:]}
for model, score in avg_cv_scores.items():
    print(f"Average Silhouette Score for {model}: {np.mean(score)}")
```

Hasil evaluasi ialah

- rata-rata nilai Silhouette Score untuk KMeans adalah sekitar 0.34,

- sedangkan untuk Hierarchical Clustering sekitar 0.33.

- Nilai Silhouette Score yang lebih tinggi menunjukkan bahwa KMeans cenderung memberikan pembagian klaster yang lebih baik daripada Hierarchical Clustering dalam dataset yang dievaluasi.

## **Tuning Hypermater**

```python
param_grid = {
    'n_clusters': [3, 4, 5, 6],  # Jumlah klaster yang akan dicoba
    'init': ['k-means++', 'random'],  # Metode inisialisasi centroid
    'algorithm': ['lloyd', 'elkan']  # Algoritma yang digunakan
}

# Evaluasi kualitas klastering menggunakan Silhouette Score
best_score = -1
best_params = None

for n_clusters in param_grid['n_clusters']:
    for init in param_grid['init']:
        for algorithm in param_grid['algorithm']:
            kmeans = KMeans(n_clusters=n_clusters, init=init, algorithm=algorithm, random_state=42, n_init='auto')
            labels = kmeans.fit_predict(X)
            silhouette_avg = silhouette_score(X, labels)
            print("Silhouette Score:", silhouette_avg, "dengan n_clusters =", n_clusters, ", init =", init, ", algorithm =", algorithm)
            if silhouette_avg > best_score:
                best_score = silhouette_avg
                best_params = {'n_clusters': n_clusters, 'init': init, 'algorithm': algorithm}

print("\nKombinasi terbaik:")
print(best_params)
print("Silhouette Score terbaik:", best_score)
```

Menunjukkan berbagai kombinasi parameter yang diuji pada model clustering menggunakan metode KMeans. Setiap baris menunjukkan kombinasi parameter yang dievaluasi, seperti jumlah klaster (n_clusters), metode inisialisasi centroid (init), dan algoritma clustering (algorithm), serta nilai Silhouette Score yang dihasilkan oleh model berdasarkan kombinasi tersebut.

Hasilnya ialah :

Informasi tentang kombinasi terbaik yang dihasilkan dari evaluasi tersebut. Kombinasi terbaik terdiri dari jumlah klaster (n_clusters) sebesar 3, metode inisialisasi centroid (init) menggunakan k-means++, dan algoritma clustering (algorithm) menggunakan 'lloyd', dengan nilai Silhouette Score terbaik sebesar 0.5848586628426434.
